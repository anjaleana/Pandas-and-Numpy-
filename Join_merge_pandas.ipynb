### Pandas Joining and Merging DataFrame

```python
import pandas as pd
```

```python
student_data1 = {'student_id':['S1','S2','S3','S4','S5'],
                 'name':['Danniella Fenton',' Ryder Storey',' Bryce Jensen','Ed Bernal',' Kwame Morin'],
                 'marks':[200,210,190,222,199]}
data = pd.DataFrame(student_data1)
data.head()
```

```python
student_data2 = {'student_id':['S1','S2','S3','S4','S5'],
                 'name':['Raju',' Ram',' Jensen','Satish','Murali'],
                 'marks':[201,200,190,222,199]}
data1 = pd.DataFrame(student_data2)
data1.head()
```

### Write a Pandas program to join the two given dataframes along rows and assign all data.

```python
result_data = pd.concat([data,data1])
result_data
```

### Write a Pandas program to join the two given dataframes along columns and assign all data

```python
result_data = pd.concat([data,data1],axis=1) # axis= 1 means column wise
result_data
```

### Write a Pandas program to append rows to an existing DataFrame and display the combined data.

```python
s6 = pd.Series(['S6', 'Scarlette Fisher', 205], index=['student_id', 'name', 'marks'])

result_data_append = result_data.append(s6,ignore_index=True)
result_data_append
```

###  Write a Pandas program to join the two dataframes using the common column of both dataframes.

```python
result_data = pd.merge(data,data1,on='student_id')
result_data
```

### Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only. 

```python
#Merged data (key from data1)
result_data_left1 = pd.merge(data,data1,how = 'left',on = ['student_id'])
print(result_data_left1)

#Merge data (key from data2)
result_data_left2 = pd.merge(data1,data,how = 'left',on = ['student_id'])
print(result_data_left2)
```

###  Write a Pandas program to join two dataframes using keys from right dataframe only

```python
result_data_right1 = pd.merge(data,data1,how = 'left',on = ['student_id'])
print(result_data_right1)

#Merge data (key from data2)
result_data_right2 = pd.merge(data1,data,how = 'left',on = ['student_id'])
print(result_data_right2)
```
